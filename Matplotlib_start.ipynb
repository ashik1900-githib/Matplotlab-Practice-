```python
import matplotlib.pyplot as plt
```

```python
lst_1 = [1,2,3,4]
lst_2 = [1,4,9,16]
plt.plot(lst_1,lst_2)
plt.plot(lst_1,lst_2, marker="o",color="Blue")
plt.title("X vs Y Comparision")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
lst_1 = [1,2,3,4]
lst_2 = [1,4,9,16]
plt.subplot(1,3,1)
plt.plot(lst_1,lst_2)
plt.plot(lst_1,lst_2,"ro")
plt.subplot(1,3,2)
plt.plot(lst_1,lst_2)
plt.plot(lst_1,lst_2,"g*")
plt.tight_layout()
plt.show()
```
